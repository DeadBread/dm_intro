# №6
Нарисуйте круговую диаграмму, показывающую соотношение способов умереть среди римских императоров.

```python
roman_emperors = pd.read_csv('data/roman-emperor-reigns.csv')

causes = roman_emperors['Cause_of_Death'].unique()

numbers = []

fig = pl.figure(figsize=(10, 10))
gr = roman_emperors.groupby('Cause_of_Death')
for name, g in gr:
    numbers.append(len(g))

pl.pie(numbers, labels=causes,
              autopct='%1.1f%%', startangle=90)
pl.show()
```
